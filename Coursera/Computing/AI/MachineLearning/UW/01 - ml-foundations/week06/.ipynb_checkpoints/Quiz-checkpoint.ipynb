```python
def dot_product(x, y):
    n = len(x)
    sum([x[i] * y[i] for i in range(n)])

def function(x):
    foo = lambda x : 1 if x else 0
    
    x1, x2, x3 = foo(x[0]), foo(x[1]), foo(x[2])
    return 1 if (x1 and x2) or ((not x1) and (not x2)) else 0

def neural_network(x, z1, z2, y):
    
    z1_in = dot_product(x, z1)
    z2_in = dot_product(x, z2)
    
    z1_in = 0 if z1_in < 0 else 1
    z2_in = 0 if z2_in < 0 else 1
    
    mid_layer = [1, z1_in, z2_in]
    
    output = dot_product(mid_layer, y)
    
    return 0 if output < 0 else 1

options = []

for number in range(8):
    bit_string = '{0:03b}'.format(number)
    options.append(map(int, list(bit_string)))
    
def test_neural_network(z1, z2, out):
    test = True
    
    for option in options:
        nn = neural_network(option, z1, z2, out)
        real = function(option)
        
        print nn, real
        
        if nn != real:
            test = False
            break
            
    if test:
        print "Ok, this NN do the job!"
    else:
        print "No, this NN sucks at this job!"
```

# Fist Neural Network

```python
z1 = [-0.5, -1, 1]
z2 = [1.5, -1, 1]
z3 = [-0.5, 1, 1]

test_neural_network(z1, z2, z3)
```

```python
z1 = [-1.5, 1, 1]
z2 = [0.5, -1, -1]
z3 = [-1.5, 1, 1]

test_neural_network(z1, z2, z3)
```
